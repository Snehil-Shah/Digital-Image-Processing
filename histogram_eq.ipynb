```python
import cv2
import matplotlib.pyplot as plt

# Step 1: Read the image
image_path = './Lenna_(test_image).png'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Check if image is loaded
if image is None:
    raise FileNotFoundError(f"Image at path '{image_path}' not found. Please check the file path.")

# Step 2: Plot histogram
plt.figure(figsize=(12, 6))

# Original Image Histogram
plt.subplot(2, 2, 1)
plt.hist(image.ravel(), 256, [0, 256])
plt.title('Original Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

# Step 3: Equalize the image
equalized_image = cv2.equalizeHist(image)

# Equalized Image Histogram
plt.subplot(2, 2, 2)
plt.hist(equalized_image.ravel(), 256, [0, 256])
plt.title('Equalized Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

# Step 4: Display the images
plt.subplot(2, 2, 3)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')  # Hide axis

plt.subplot(2, 2, 4)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')  # Hide axis

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Read the image
image_path = 'Lenna_(test_image).png'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Check if image is loaded
if image is None:
    raise FileNotFoundError(f"Image at path '{image_path}' not found. Please check the file path.")

# Step 2: Perform bit slicing
bit_planes = []
for i in range(8):
    bit_plane = (image >> i) & 1  # Extract the i-th bit plane
    bit_planes.append(bit_plane * 255)  # Scale to 0-255 for visualization

# Step 3: Display the original image and the bit planes
plt.figure(figsize=(12, 8))

# Display original image
plt.subplot(3, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Display each bit plane
for i in range(8):
    plt.subplot(3, 3, i + 2)
    plt.imshow(bit_planes[i], cmap='gray')
    plt.title(f'Bit Plane {i}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
