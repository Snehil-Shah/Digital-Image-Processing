# Exploring CV techniques

```python
import cv2
import numpy as np

# Read the image
img = cv2.imread('Lenna_(test_image).png')

# Split the image into R, G, B channels
b, g, r = cv2.split(img)

# Create two empty layers (black images)
empty1 = np.zeros(img.shape[:2], dtype=np.uint8)
empty2 = np.zeros(img.shape[:2], dtype=np.uint8)

# Create 3-channel images for each color and the empty layers
red_channel = cv2.merge([empty1, empty2, r])
green_channel = cv2.merge([empty1, g, empty2])
blue_channel = cv2.merge([b, empty2, empty1])


# Stack the images horizontally
result = np.hstack((img, red_channel, green_channel, blue_channel))

import matplotlib.pyplot as plt

# Display the images
plt.figure(figsize=(10, 8))
plt.imshow(result)
plt.axis('off')
plt.show()
```

```python

# Convert the RGB layers to grayscale using different methods
image = cv2.imread('lena.png')
r, g, b = cv2.split(image)


lightness = (np.min(image, axis=2) + np.max(image, axis=2)) / 2
average = (r + g + b) / 3
# Luminosity method
luminosity = 0.3 * r + 0.59 * g + 0.11 * b


lightness = np.uint8(lightness)
average = np.uint8(average)
luminosity = np.uint8(luminosity)

# Display the grayscale images
plt.figure(figsize=(10, 8))
plt.subplot(131), plt.imshow(lightness, cmap='gray'), plt.title('Lightness')
plt.subplot(132), plt.imshow(average, cmap='gray'), plt.title('Average')
plt.subplot(133), plt.imshow(luminosity, cmap='gray'), plt.title('Luminosity')

plt.show()



```
